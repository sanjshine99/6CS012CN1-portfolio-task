```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Get some data
x = np.array([2.5, 0.5, 2.2, 1.9, 3.1,2.3,2,1,1.5,1.1])
y = np.array([2.4, 0.7, 2.9, 2.2, 3.0,2.7,1.6,1.1,1.6,0.9])

# Combine x and y into a 2D array
data = np.column_stack((x, y))

# Step 2: Calculate the mean and subtract from x and y
mean_data = np.mean(data, axis=0)
centered_data = data - mean_data

# Step 3: Calculate the covariance matrix
covariance_matrix = np.cov(centered_data, rowvar=False)

# Step 4: Calculate the eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Step 5: Choosing components and forming a feature vector
# In this example, we choose all components since the data is 2D
feature_vector = eigenvectors

# Step 6: Deriving the new data set
new_data = np.dot(centered_data, feature_vector)

# Getting the original data back
reconstructed_data = np.dot(new_data, feature_vector.T) + mean_data

# Print results
print("Original Data:")
print(data)
print("\nMean-Centered Data:")
print(centered_data)
print("\nCovariance Matrix:")
print(covariance_matrix)
print("\nEigenvectors:")
print(eigenvectors)
print("\nEigenvalues:")
print(eigenvalues)
print("\nFeature Vector:")
print(feature_vector)
print("\nNew Data Set:")
print(new_data)
print("\nReconstructed Data:")
print(reconstructed_data)

# Plot the original and reconstructed data
plt.scatter(x, y, label='Original Data')
plt.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], label='Reconstructed Data', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('PCA: Original and Reconstructed Data')
plt.show()
```
